```python
from datetime import datetime, timedelta

#日付だけのデータフレームを作成
date_list = [datetime(2020, 1, 1, hour=0, minute=0, second=0) + timedelta(days=i) for i in range(397)]
df_index_date = pd.DataFrame(date_list, columns = ["date"])
#print(date_str_list)
display(df_index_date)
```

```python
df_index_date['date'] = df_index_date["date"].apply(lambda w: pd.to_datetime(w))
df_index_date
```

```python
# 結合用コード

#コロナ
columns_list = df_indicator_csv_1.columns.values

for c in range(1, len(columns_list)):
    df_index_date[columns_list[c]] = ''

for i in range(len(df_indicator_csv_1)):
#for i in range(30):
    for j in range(len(df_index_date)):
    #for j in range(30):
        if df_indicator_csv_1.iloc[i]["日付"] == df_index_date.iloc[j]["date"]:
            for f in range(1, len(columns_list)):
                df_index_date.loc[j, columns_list[f]] = df_indicator_csv_1.loc[i, columns_list[f]]

display(df_index_date.head(30))
display(df_index_date.tail(30))
```
